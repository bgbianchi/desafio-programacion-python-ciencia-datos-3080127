```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('02_04_datos.csv', delimiter=";")
```

```python
print(df.head(20))
```

```python
scaler = MinMaxScaler()
```

- Normalización: Implica ajustar datos enteros que existen en diferentes escalas en una escala común para poder compararlos.
- El MinMaxScaler es una técnica de escalado (scaling) de datos. Su función principal es transformar los valores numéricos de una característica (columna) para que se encuentren dentro de un rango específico, generalmente entre 0 y 1.
    - MinMaxScaler funciona mediante una transformación lineal simple. Reajusta los datos de manera que el valor mínimo original de la característica se convierte en 0 y el valor máximo original se convierte en 1. Todos los valores intermedios se escalan proporcionalmente dentro de ese rango \([0,1]\)


- Desafío: Aplicar una técnica de normalización en las columnas del dataframe

```python
columnas_a_escalar = ['Altura', 'Peso']
columnas_escaladas = scaler.fit_transform(df[columnas_a_escalar])
df_escalado = pd.DataFrame(columnas_escaladas, columns=columnas_a_escalar)
print(df_escalado.head(20))
```
