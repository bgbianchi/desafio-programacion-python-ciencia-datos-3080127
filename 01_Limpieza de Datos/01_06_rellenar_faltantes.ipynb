- Desafío: Llenar datos faltantes de la columna edad con la media de edad

```python
# By default, .mean() automatically skips NaN values
edad_media = df['Edad'].mean()
print(edad_media)
```

```python
df.query("Edad.isnull()")
```

```python
df.loc[df["Edad"].isnull(), "Edad"] = edad_media
print(df.head(10))
```

```python
df.query("Edad.isnull()")
```

```python
# Solución Alternativa
df["Edad"] = df["Edad"].fillna(edad_media)
print(df.head(10))
print(df.isnull().sum())
```
