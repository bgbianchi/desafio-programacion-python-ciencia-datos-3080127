```python
import pandas as pd
```

```python
df = pd.read_csv('01_05_empleados.csv', delimiter=";")
```

```python
print(df.shape)
```

```python
duplicados = df.duplicated()
```

```python
print(df[duplicados])
```

- Desafío: Eliminar duplicaciones en el dataframe

```python
df_sin_duplicados = df.drop(df[duplicados].index)
print(df_sin_duplicados.duplicated().sum())
print(df_sin_duplicados.shape)
```

```python
df_sin_duplicados = df.drop_duplicates()
print(df_sin_duplicados.duplicated().sum())
print(df_sin_duplicados.shape)
```
