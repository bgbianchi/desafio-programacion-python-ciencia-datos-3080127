```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
X, y = make_classification(n_samples=5000,  n_features=20, random_state=3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

- Random Forest: Algoritmo de aprendizaje automático conjunto, y un conjunto de algoritmos de árboles de decisión.
    - Tiene excelente rendimiento en modelado predictivo de clasificación y regresión.

- Desafío: 
    1. Crear un modelo de clasificación usando RandomForestClassifier, establecer random_state en 42
    2. Entrenar el modelo
    3. Imprimir la precisión del modelo
    4. Realizar predicciones con los datos de prueba
    5. Graficar la matriz de confusión

```python
# Hiperparámetros clave:
# n_estimators: El número de árboles en el bosque (cuantos más, mejor rendimiento, pero más lento)
# criterion: 'gini' o 'entropy' (gini es el predeterminado)
# random_state: Para asegurar resultados reproducibles
model = RandomForestClassifier(n_estimators=100, criterion='gini', random_state=42)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"\nPrecisión (Accuracy) del modelo: {accuracy:.2f}")
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm, 
    annot=True,          # Añadir los números dentro de los cuadros
    fmt="d",             # Formato de dígito (sin notación científica)
    cmap="Blues",        # Mapa de colores (puedes usar 'viridis', 'plasma', etc.)
)

plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusión')
plt.show()
```
