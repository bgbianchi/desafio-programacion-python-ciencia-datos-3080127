```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv('04_07_empleados.csv', delimiter=";")
```

```python
print(df.head())
```

- Bosque de aislamiento:
    - Algoritmo de aprendizaje automático no supervisado que identifica anomalías

- Desafío: Detectar aquellos salarios que sean atípicos
    1. Definir y ajustar el modelo usando IsolationForest con un random_state en 42
    2. Agregar una nueva columna al dataframe denominada "Normal"
    3. Asignar a la columna "Normal"  los valores devueltos por la función predict() pasando el salario como parámetro
    4. Encontra la cantidad de valores atípicos e imprimir el resultado en consola

```python
# n_estimators: Número de árboles a construir (más es mejor, 100 es común)
# contamination: La proporción esperada de outliers (ej. 10%)
# random_state: Para resultados reproducibles
model = IsolationForest(n_estimators=100, contamination='auto', random_state=42)
model.fit(df[['Salario']])

df['Normal'] = model.predict(df[['Salario']])

print(df['Normal'].value_counts())
```
